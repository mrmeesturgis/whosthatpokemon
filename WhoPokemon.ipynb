```python
import tkinter as tk
import tkinter.ttk as ttk
from PIL import ImageTk, Image
import random
```

```python
class W_Pokemon():
    def __init__(self):
        self.max = 16
        self.grif = tk.Tk()
        self.imgs = self.iloader()
        self.counter = 0
        self.c2 = 0
        self.targets = {}
        self.pokeball = ImageTk.PhotoImage(Image.open('images/pball.jpeg').resize((200,200)))
        for i, img in enumerate(self.imgs):
            self.imgs[i] = self.pixelator(img)            
        #self.test = self.labeler()
        #self.updater()
        self.pokegrid()
        self.grif.mainloop()
    
    def iloader(self):
        imgs = []
        for i in range(1, self.max):
            base = Image.open(f"images/p{i}.jpeg")
            final = base.resize((200,200))
            imgs.append(final)
        random.shuffle(imgs)
        return imgs
        
    def pixelator(self, target):
        #Load the image for manipulation
        base = target.resize((200,200))
        nf = [200, 100, 50, 40, 20, 15, 12, 10, 8, 5, 2]
        nf.reverse()
        print(nf)
        #Divide into cells according to frames
        sequence = []
        for frame in nf:
            ftarget = base
            pixels = ftarget.load()
            counter = 0
            base_row = 0
            cell = {}
            for reach_row in range(0, ftarget.size[0] +1):
                if reach_row % frame == 0 and reach_row != 0: #Need a list of numbers for the modulus
                    for i in range(base_row,reach_row):
                        base_col = 0
                        for reach_col in range(0, ftarget.size[1]+1):
                            if reach_col % frame == 0 and reach_col != 0:
                                try: 
                                    cell[((f"{base_row}",f"{base_col}"),(f"{reach_row}",f"{reach_col}"))]
                                except:
                                    cell[((f"{base_row}",f"{base_col}"),(f"{reach_row}",f"{reach_col}"))] = []
                                for j in range(base_col,reach_col):
                                    cell[((f"{base_row}",f"{base_col}"),(f"{reach_row}",f"{reach_col}"))].append(((i,j),pixels[i,j]))
                                base_col = reach_col
                    base_row = reach_row

            #Pixelate
            #Find averages
            rewrite = {}
            for k, v in cell.items():
                r = sum(x[1][0] for x in v)//len(v)
                g = sum(x[1][1] for x in v)//len(v)
                b = sum(x[1][2] for x in v)//len(v)
                for pix in v:
                    rewrite[pix[0]] = (r,g,b)
            #Use rewrite
            for k, v in rewrite.items():
                pixels[k[0],k[1]] = v
            sequence.append(ImageTk.PhotoImage(ftarget.resize((200,200))))
        
        #Outside of the loop, add the last image
        
        sequence.reverse()
        sequence.append(ImageTk.PhotoImage(target.resize((200,200))))
        return sequence
    
    
    def labeler(self):
        test = tk.Label(image = self.imgs[0][0])
        test.pack()
        return test
        
    def pokegrid(self):
        buttons = []
        targets = {}
        txt = []
        counter = 0
        for i in range(0,2):
            self.grif.columnconfigure(i, weight=1, minsize=75)
            self.grif.rowconfigure(i, weight=1, minsize=50)

            for j in range(0, 2):
                self.counter += 1
                frame = tk.Frame(
                    master=self.grif,
                    relief=tk.RAISED,
                    borderwidth=1
                )
                frame.grid(row=i, column=j, padx=5, pady=5)
                buttons.append(tk.Button(master = frame, image = self.pokeball, command = lambda x= f"{i}"+f"{j}": self.morph(x)))
                buttons[-1].pack()
                self.targets[f"{i}"+f"{j}"]=buttons[-1]
    
    def updater(self, dynamic, refer):
        self.c2 += 1
        try:
            dynamic['image'] = refer[self.c2]
            self.grif.after(500, self.updater, dynamic, refer)
        except:
            self.c2 = 0
            dynamic['image'] = self.pokeball
            
        #print(str(img))


    
    def morph(self, x):
        y = random.randint(0, len(self.imgs)-1)
        #First to produce an image
        if str(self.targets[x]["image"]) == str(self.pokeball):
            self.targets[x]["image"] = f"{self.imgs[y][0]}"
            self.updater(self.targets[x], self.imgs[y])
        #Second to return to a pokeball
        else:
            self.targets[x]["image"] = self.pokeball
```
